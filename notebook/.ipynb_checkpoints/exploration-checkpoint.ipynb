```python
import pandas as pd
import numpy as np

# Read the file manually line by line to handle the format
rows = []
with open('/Users/rorypius/Desktop/fuel-efficiency-predictor/data/auto-mpg.data', 'r') as f:
    for line in f:
        line = line.strip()
        if not line:
            continue
        # Split by whitespace but keep quoted strings together
        import shlex
        try:
            parts = shlex.split(line)
            if len(parts) >= 8:
                rows.append(parts[:8])  # Take only first 8 values
        except:
            continue

columns = ['mpg','cylinders','displacement','horsepower',
           'weight','acceleration','model_year','origin']

df = pd.DataFrame(rows, columns=columns)

# Convert all columns to numbers
for col in columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Replace ? with NaN and drop missing
df = df.replace('?', np.nan).dropna()

print(df.head())
print('Shape:', df.shape)
```

```python
import pandas as pd
import numpy as np

# Read the file manually line by line to handle the format
rows = []
with open('/Users/rorypius/Desktop/fuel-efficiency-predictor/data/auto-mpg.data', 'r') as f:
    for line in f:
        line = line.strip()
        if not line:
            continue
        # Split by whitespace but keep quoted strings together
        import shlex
        try:
            parts = shlex.split(line)
            if len(parts) >= 8:
                rows.append(parts[:8])  # Take only first 8 values
        except:
            continue

columns = ['mpg','cylinders','displacement','horsepower',
           'weight','acceleration','model_year','origin']

df = pd.DataFrame(rows, columns=columns)

# Convert all columns to numbers
for col in columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Replace ? with NaN and drop missing
df = df.replace('?', np.nan).dropna()

print(df.head())
print('Shape:', df.shape)
```

```python
# Statistical summary of every column
df.describe()
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt

# Plot 1 — Distribution of MPG
plt.figure(figsize=(8, 4))
plt.hist(df['mpg'], bins=20, color='#00d4ff', edgecolor='black')
plt.title('Distribution of Fuel Efficiency (MPG)')
plt.xlabel('Miles Per Gallon')
plt.ylabel('Number of Cars')
plt.show()
```

```python
# Plot 2 — Engine Size vs Fuel Efficiency
plt.figure(figsize=(8, 4))
plt.scatter(df['displacement'], df['mpg'], 
            alpha=0.5, color='#7c3aed')
plt.title('Engine Size vs Fuel Efficiency')
plt.xlabel('Displacement (Engine Size)')
plt.ylabel('Miles Per Gallon')
plt.show()
```

```python
# Plot 3 — Average MPG by Year
df.groupby('model_year')['mpg'].mean().plot(
    kind='line', 
    color='#00e5a0', 
    marker='o',
    figsize=(8, 4)
)
plt.title('Average Fuel Efficiency by Year (1970-1982)')
plt.xlabel('Year')
plt.ylabel('Average MPG')
plt.show()
```
